```python
df_age_dist = pd.read_sql_query("SELECT native_country, SUM(CASE WHEN age < 18 THEN 1 ELSE 0 END) AS [Under 18]," +
       " SUM(CASE WHEN age BETWEEN 18 AND 24 THEN 1 ELSE 0 END) AS [18-24],"+
       " SUM(CASE WHEN age BETWEEN 25 AND 34 THEN 1 ELSE 0 END) AS [25-34], "+
       " SUM(CASE WHEN age BETWEEN 35 AND 45 THEN 1 ELSE 0 END) AS [35-45], "+
       " SUM(CASE WHEN age BETWEEN 46 AND 60 THEN 1 ELSE 0 END) AS [46-60], "+     
       " SUM(CASE WHEN age BETWEEN 61 AND 75 THEN 1 ELSE 0 END) AS [61-75], "+
       " SUM(CASE WHEN age BETWEEN 76 AND 100 THEN 1 ELSE 0 END) AS [76-100] "+
 " FROM sqladb "+
   " group by native_country order by native_country", conn)
print(df_age_dist)
```
